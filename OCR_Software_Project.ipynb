```python
from paddleocr import PaddleOCR, draw_ocr
from google.colab.patches import cv2_imshow
import cv2

ocr = PaddleOCR(use_angle_cls=True, use_gpu=False, lang='en', show_log=False)
img_path = 'HW_Text-2.jpg'
img = cv2.imread(img_path, cv2.COLOR_BGR2RGB)
cv2_imshow(img)

result = ocr.ocr(img_path)

print('\n\nThe predicted output is:\n')
i = 0
for res in result:
  while(i<len(result[0])):
    print(f'{res[i][1][0]} _____________________ Accuracy: {round(res[i][1][1]*100,2)} %')
    i += 1
```
