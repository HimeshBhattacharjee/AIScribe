```python
%%capture
%%writefile AIScribe.py
import streamlit as st
from paddleocr import PaddleOCR

ocr = PaddleOCR(use_angle_cls=True, use_gpu=False, lang='en', show_log=False)

# Title of the web app
st.title(':red[AI]Scribe')

# Header
st.header('Handwritten :blue[Digit] and :green[Character] Recognition')

# Displaying a file uploader
file = st.file_uploader('Upload an image', type=["png", "jpg", "jpeg"])

# Displaying an image
if file is not None:
    st.image(file, caption='Uploaded image', use_column_width='auto')
    img = file.read()

# Displaying a button
if st.button('Predict'):
    # Displaying a progress bar
    with st.spinner('Predicting...'):
        import time
        time.sleep(2)
        st.success('OCR Successful!')

    result = ocr.ocr(img, cls=False)
    if result[0] is not None:
        st.write('\n\nThe predicted output is:\n')
        i = 0
        for res in result:
            while(i<len(result[0])):
                st.write(f'{res[i][1][0]} _____________________ Accuracy: {round(res[i][1][1]*100,2)} %')
                i += 1
    else:
        result = ocr.ocr(img, det=False, cls=False)
        st.write('\n\nThe predicted output is:\n')
        i = 0
        for res in result:
            while(i<len(result[0])):
                st.write(f'{res[i][0]} _____________________ Accuracy: {round(res[i][1]*100,2)} %')
                i += 1
```

```python
import os
from threading import Thread
from pyngrok import ngrok
def run_streamlit():
    os.system('streamlit run AIScribe.py --server.port 8500')
# Start a thread to run the Streamlit app
thread = Thread(target=run_streamlit)
thread.start()
```
